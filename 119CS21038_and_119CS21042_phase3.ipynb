```python
import pandas as pd
df=pd.read_csv('crimes_against_women_2001-2014.csv')
df
```

```python
# Calculate descriptive statistics for a specific column
rape_stats = df['Rape'].describe()
print(rape_stats)
```

```python
import matplotlib.pyplot as plt

# Plot a histogram for 'Rape' cases
plt.hist(df['Rape'], bins=20, color='blue', alpha=0.7)
plt.title('Histogram of Rape Cases')
plt.xlabel('Number of Cases')
plt.ylabel('Frequency')
plt.show()
```

```python
# Filter the data for the year 2014
data_2014 = df[df['Year'] == 2014]

# Group data by state and calculate the sum of 'Rape' cases for each state
state_rape_cases_2014 = data_2014.groupby('STATE/UT')['Rape'].sum().reset_index()

# Sort the data by the number of cases for better visualization
state_rape_cases_2014 = state_rape_cases_2014.sort_values(by='Rape', ascending=False)

# Create a bar plot
plt.figure(figsize=(10, 6))
plt.barh(state_rape_cases_2014['STATE/UT'], state_rape_cases_2014['Rape'], color='skyblue')
plt.title('Number of Reported Rape Cases by State/UT (2014)')
plt.xlabel('Number of Cases')
plt.ylabel('State/UT')
plt.grid(axis='x')
plt.show()
```

```python
import seaborn as sns

# Create a box plot for 'Rape' cases
sns.boxplot(x=df['Rape'], color='green')
plt.title('Box Plot of Rape Cases')
plt.xlabel('Rape Cases')
plt.xticks(rotation=70)
plt.show()
```

```python
# Create a bar chart for the 'STATE/UT' column
state_counts = df['STATE/UT'].value_counts()
state_counts.plot(kind='bar', color='purple')
plt.title('Bar Chart of States/UT')
plt.xlabel('State/UT')
plt.ylabel('Frequency')
plt.show()
```

  **2**.**Bivariate** **Analysis**

```python
import matplotlib.pyplot as plt

# Create a scatter plot to analyze the relationship between 'Rape' and 'Kidnapping and Abduction'
plt.scatter(df['Rape'], df['Kidnapping and Abduction'], color='blue', alpha=0.7)
plt.title('Scatter Plot: Rape vs. Kidnapping and Abduction')
plt.xlabel('Rape Cases')
plt.ylabel('Kidnapping and Abduction Cases')
plt.show()
```

```python
# Calculate the Pearson correlation coefficient between 'Rape' and 'Kidnapping and Abduction'
correlation = df['Rape'].corr(df['Kidnapping and Abduction'])
print("Pearson Correlation Coefficient:", correlation)
```

```python
import seaborn as sns

# Create a box plot to compare the distribution of 'Rape' cases across states
sns.boxplot(x=df['STATE/UT'], y=df['Rape'], color='green')
plt.title('Box Plot: Rape Cases by State/UT')
plt.xlabel('State/UT')
plt.ylabel('Rape Cases')
plt.xticks(rotation=90)
plt.show()
```

```python
# Create a cross-tabulation between 'STATE/UT' and 'Year' to analyze the distribution of states over the years
cross_tab = pd.crosstab(df['STATE/UT'], df['Year'])
print(cross_tab)
```

**3**.**Multivariate** **Analysis**

```python
from sklearn.decomposition import PCA

# Perform PCA on the dataset to reduce dimensionality
X = df[['Rape', 'Kidnapping and Abduction', 'Dowry Deaths', 'Assault on women with intent to outrage her modesty']]
pca = PCA(n_components=2)
principal_components = pca.fit_transform(X)
explained_variance = pca.explained_variance_ratio_
print("Explained Variance:", explained_variance)

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select a subset of columns for demonstration
subset = df[['Rape', 'Kidnapping and Abduction', 'Dowry Deaths', 'Assault on women with intent to outrage her modesty']]

# Add 'Year' as a categorical variable for coloring
subset['Year'] = df['Year'].astype('category')

# Create a scatterplot matrix
sns.set(style="ticks")
sns.pairplot(subset, hue='Year', palette="husl")
plt.show()
```

```python
# Calculate the correlation matrix
correlation_matrix = subset.corr()

# Create a correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Group data by year and calculate the sum of 'Rape' cases for each year
yearly_rape_cases = df.groupby('Year')['Rape'].sum()

# Plot the line chart
plt.figure(figsize=(10, 6))
plt.plot(yearly_rape_cases.index, yearly_rape_cases.values, marker='o', linestyle='-')
plt.title('Trend of Reported Rape Cases (2001-2014)')
plt.xlabel('Year')
plt.ylabel('Number of Cases')
plt.grid(True)
plt.xticks(yearly_rape_cases.index, rotation=45)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a FacetGrid based on the 'STATE/UT' column
g = sns.FacetGrid(df, col="STATE/UT", col_wrap=4, height=3, aspect=1.5)
g.map(sns.lineplot, "Year", "Rape", marker="o", ci=None, palette="husl")
g.set_axis_labels("Year", "Number of Rape Cases")
g.set_titles(col_template="{col_name}")

# Rotate the x-axis labels for better visibility
plt.xticks(rotation=45)
plt.subplots_adjust(top=0.9)
g.fig.suptitle("Trend of Reported Rape Cases by State/UT (2001-2014)", fontsize=16)
plt.show()
```

```python
# Filter the data for the year 2014
data_2014 = df[df['Year'] == 2014]

# Create a box plot
plt.figure(figsize=(12, 6))
ax = sns.boxplot(x='STATE/UT', y='Rape', data=data_2014, palette="Set2")
plt.xticks(rotation=90)
plt.title('Box Plot of Reported Rape Cases by State/UT (2014)')
plt.xlabel('State/UT')
plt.ylabel('Number of Cases')
plt.show()
```
